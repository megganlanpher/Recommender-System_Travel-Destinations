```python
country = 'Botswana'

import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import table
from gensim.models import Doc2Vec


model= Doc2Vec.load("d2v_model.bin")

recommendations = model.docvecs.most_similar(country, topn=50)
# recommendations

df_recomm = pd.DataFrame(recommendations, columns=['Country', 'Similarity'])
ax = plt.subplot(50,1,1, frame_on=False)
ax.xaxis.set_visible(False) 
ax.yaxis.set_visible(False) 

table(ax, df_recomm);
```
